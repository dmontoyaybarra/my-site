```python
import yahoo_fin.stock_info as yf
import datetime as dt
import yfinance as yff
import plotly.express as px 
import pandas as pd
```

```python
start = dt.datetime(2021,1,1)
end = dt.datetime(2021,12,12)
```

```python
yf.get_dividends('AAPL',start_date=start,end_date=end)
```

```python
def graph(ticker_company):
    stock_data = yff.Ticker(ticker_company).history("5y")
    return px.line(stock_data['Close'])
    
```

```python
graph("aapl")
```

```python
stock_data = yff.Ticker("MSFT").history("5y")
px.line(stock_data['Close'])
```
